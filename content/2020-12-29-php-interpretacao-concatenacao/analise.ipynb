# Interpretação de Variáveis em String ou Concatenação?

```r
dados <- read.csv('dados.csv.gz', colClasses=c('factor', 'numeric', 'integer'))
```

## Tempo de Execução

```r
boxplot(
    tempo ~ codigo,
    dados,
    main = 'Tempos de Execução dos Códigos',
    xlab = 'Códigos',
    ylab = 'Tempo de Execução (em segundos)',
    col = '#8ebed3'
)
```

```r
summary(dados.aov_tempo <- aov(tempo ~ codigo, dados))
```

```r
plot(TukeyHSD(dados.aov_tempo))
```

## Uso de Memória

```r
boxplot(
    memoria ~ codigo,
    dados,
    main = 'Memória Usada pelos Códigos',
    xlab = 'Códigos',
    ylab = 'Total de Memória (em KB)',
    col = '#8ebed3'
)
```

```r
summary(dados.aov_memoria <- aov(memoria ~ codigo, dados))
```

```r
plot(TukeyHSD(dados.aov_memoria))
```
